```python
df["Ticket_Price_Cr"]=df["Ticket_Price_Cr"].str.replace('₹','')
df["Ticket_Price_Cr"]=df["Ticket_Price_Cr"].str.replace('Cr','')
df["Ticket_Price_Cr"]=df["Ticket_Price_Cr"].str.replace(' ','')
df.head(10)
```

```python
df["Purchase_Quarter"]=pd.to_datetime(df["Purchase_Quarter"])
```

```python
df
```

```python
df.dtypes
```

```python
df["Ticket_Price_Cr"]=pd.to_numeric(df["Ticket_Price_Cr"],errors='coerce')
```

```python
df["Ticket_Price_Cr"]=df["Ticket_Price_Cr"].round(2)
```

```python
df
```

```python
df["Micro_Market"]=df["Micro_Market"].str.title()
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df["Buyer_Comments"].fillna("No comments",inplace=True)
```

```python

df["Amenity_Score"] = pd.to_numeric(df["Amenity_Score"], errors='coerce')
df["Amenity_Score"]=df["Amenity_Score"].fillna(df["Amenity_Score"].median())
```

```python
df.isnull().sum()
```

```python
df.loc[df["Unit_Size_Sqft"] <= 0,"Unit_Size_Sqft"]=np.nan
df.loc[df["Ticket_Price_Cr"] <= 0,"Ticket_Price_Cr"]=np.nan
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Ticket_Price_Cr"]=df["Ticket_Price_Cr"].fillna(df["Ticket_Price_Cr"].median())
```

```python
df["Unit_Size_Sqft"]=df["Unit_Size_Sqft"].fillna(df["Unit_Size_Sqft"].median())
```

```python
df["Ticket_Price_Cr"].isna().sum()
```

```python
df["Developer_Name"]=df["Developer_Name"].str.title()
```

```python
df["Price_Per_Sqft"]=(df["Ticket_Price_Cr"]*10000000/df["Unit_Size_Sqft"]).round()
```

```python
df["Year"]=df["Purchase_Quarter"].dt.year
```

```python
df["Quarter"]=df["Purchase_Quarter"].dt.quarter
```
